# **Jobs Cleaned**

```python
jobs_cleaned = jobs.drop(['Competitors','Easy Apply', 'Founded'], axis=1).copy()
```

```python
jobs_cleaned
```

# Rows having nulls

```python
jobs_cleaned.replace([-1,'-1'],np.nan, inplace=True)
```

```python
jobs_cleaned.isna().sum()
```

```python
# rows_having_nulls = jobs_cleaned[jobs_cleaned.eq(-1).any(axis=1)]
```

```python
jobs_cleaned
```

```python
jobs_cleaned['Salary Estimate'] = jobs_cleaned['Salary Estimate'].str.split(' ').str[0]
jobs_cleaned['Salary Estimate'] = jobs_cleaned['Salary Estimate'].str.strip()
jobs_cleaned['Salary Estimate'] = jobs_cleaned['Salary Estimate'].str.replace('$','')
```

```python
x = jobs_cleaned.copy()
```

```python
jobs_cleaned[['min_salary','max_salary']] = jobs_cleaned['Salary Estimate'].str.split('-',expand=True)
```

```python
def clean_sal_col(z, dataset):
    dataset[z] = dataset[z].replace('', np.nan)  # there was a row where x['a'] was empty
    dataset[z] = dataset[z].str.replace('K','', case=False).str.strip()
    dataset[z] = dataset[z].astype(float)
    dataset[z] = dataset[z]*1000  # as we had removed K
    dataset[z].fillna(0)
    return dataset
```

```python
clean_sal_col('max_salary', jobs_cleaned)
```

```python
clean_sal_col('min_salary',  jobs_cleaned)
```

```python
jobs_cleaned['avg_salary'] = jobs_cleaned[['min_salary','max_salary']].mean(axis=1)         # apllying mean on horizontally
```

```python
jobs_cleaned = jobs_cleaned.dropna(subset=['min_salary','max_salary'])
```
